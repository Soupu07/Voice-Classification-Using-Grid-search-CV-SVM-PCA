```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('voice-classification.csv')
```

```python
df.head()
```

```python
df.info()
```

# Lets pickup the IV and DV

```python
X = df.iloc[:,0:-1]
X.info()
```

```python
y = df.iloc[:, -1]
y
```

```python
from sklearn.preprocessing import LabelEncoder
gender_encoder = LabelEncoder()
```

```python
y = gender_encoder.fit_transform(y)
y
```

```python
# SVM is distance based ML algo , we awill apply standard scaler

#from sklearn.preprocessing import StandardScaler

#scaler = StandardScaler()
```

```python
#X = scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state= 123)
```

# Applying the Suuport vector classifier

```python
from sklearn.svm import SVC
from sklearn import metrics
```

```python
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
y_pred = svc_model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

# Model Evaluation

```python
accuracy_svm = metrics.accuracy_score(y_test, y_pred)
accuracy_svm
```

```python
# Lets print the confusion matrix 

print(metrics.confusion_matrix(y_test, y_pred))
```

```python
# Lets see if my model is overfitted

y_pred_train = svc_model.predict(X_train)
print(metrics.accuracy_score(y_train, y_pred_train))
```

# lets do the hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[.1,1,10, 100], 'gamma': [100, 10, 1, .1, .01, .001 ]}
```

```python
param_grid
```

```python
grid = GridSearchCV(SVC(), param_grid, refit= True, verbose = 2)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid_prediction = grid.predict(X_test)
```

```python
grid_prediction
```

```python
print(metrics.accuracy_score(y_test,grid_prediction))
```

```python
grid.best_params_
```

```python
# Lets create the final model with the help of the final and best parameters found ot in the Gridsearch CV
```

```python
svc_grid_model = SVC(C=1.0 , gamma = 0.1)
```

```python
svc_grid_model
```

```python
svc_grid_model.fit(X_train , y_train)
```

```python
y_grid_pred = svc_grid_model.predict(X_test)
y_grid_pred
```

```python
print(metrics.accuracy_score(y_test , y_grid_pred))
```
